## 2. Bipartite graphs

We now consider a large bipartite graph. We consider the articles-words bipartite graph of the vital articles of WikiVitals.

## Wikipedia Vitals

```python
graph = wikivitals
```

```python
biadjacency = graph.biadjacency
articles = graph.names
words = graph.names_col
labels = graph.labels
categories = graph.names_labels
```

```python
def get_neighbors(biadjacency, i):
    return biadjacency.indices[biadjacency.indptr[i]: biadjacency.indptr[i + 1]]
```

```python
n = len(articles)
```

```python
sample = np.random.choice(n, size=10)
```

```python
print(articles[sample])
```

```python
print(labels[sample])
```

```python
print(categories[labels[sample]])
```

```python
i = sample[0]
print(articles[i])
```

```python
neighbors = get_neighbors(biadjacency, i)
print(words[neighbors])
```

```python
labels_unique, counts_categories = np.unique(labels, return_counts=True)
```

```python
for l in labels_unique:
    print(categories[l], counts_categories[l])
```
