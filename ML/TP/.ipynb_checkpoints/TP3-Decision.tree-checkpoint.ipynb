QUESTION 2

Simulez avec rand_checkers des échantillons de taille n = 456 (attention à bien équilibrer les
classes). Créez deux courbes qui donnent le pourcentage d’erreurs commises en fonction de la
profondeur maximale de l’arbre (une courbe pour Gini, une courbe pour l’entropie). On laissera les
autres paramètres à leur valeurs par défaut.

```python
data5 = rand_checkers(n1=114, n2=114, n3=114, n4=114, sigma=0.1)
X5 = data5[:,0:2]
y5 = data5[:,-1]
```

```python
# gini
clf_gini = tree.DecisionTreeClassifier(criterion='gini',max_depth=None)
clf_gini.fit(X5,y5)
clf_gini.predict(X5)

# taux d'erreur
gini_errors = 1- clf_gini.score(X5,y5)

print(f"paramètre du modle {clf_gini.get_params()}")
print(f"gini_errors  {gini_errors }")
```

```python
# entropy

clf_entropy = tree.DecisionTreeClassifier(criterion='entropy',max_depth=None)
clf_entropy .fit(X5,y5)
clf_entropy .predict(X5)

# taux d'erreur
entropy_errors = 1-clf_entropy .score(X5,y5)


print(f"paramètre du modle {clf_entropy.get_params()}")
print(f"entropy_errors {entropy_errors}")
```

```python
# creation des courbes d'erreur en fonction de la profondeur maximale
idmax =20
max_depths = range(idmax)
critere = ['gini','entropy']
for criterions in critere:
    scores=[]
    for i in max_depths :
        clf = tree.DecisionTreeClassifier(criterion=criterions,max_depth=i+1)
        clf.fit(X5,y5)
        scores.append((1-clf.score(X5,y5))*100)
    plt.plot(max_depths,scores)
    plt.title(criterions)
    plt.show()
    
```

```python
# The digits dataset
digits = datasets.load_digits()

# To apply a classifier on this data, we need to flatten the image, to
# turn the data in a (samples, feature) matrix:
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))

# Split data into train and test subsets
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.2, shuffle=False)

# Create a classifier: a support vector classifier
# creation des courbes d'erreur en fonction de la profondeur maximale
idmax =20
max_depths = range(idmax)
critere = ['gini','entropy']
for criterions in critere:
    scores=[]
    for i in max_depths :
        clf7 = tree.DecisionTreeClassifier(criterion=criterions,max_depth=i+1)
        clf7.fit(X_train,y_train)
        scores.append((1-clf7.score(X_test,y_test))*100)
    plt.plot(max_depths,scores)
    plt.title(criterions)
    plt.show()
```

```python
# gini
clf7_gini = tree.DecisionTreeClassifier(criterion='gini',max_depth=None)
clf7_gini.fit(X_train,y_train)

# taux d'erreur
gini_errors = 1-clf7_gini.score(X_test,y_test)

print(f"paramètre du modele {clf7_gini.get_params()}")
print(f"gini_errors {gini_errors}")



# entropy
clf7_entropy = tree.DecisionTreeClassifier(criterion='entropy',max_depth=None)
clf7_entropy.fit(X_train,y_train)

# taux d'erreur
entropy_errors = 1-clf7_entropy.score(X_test,y_test)

print(f"paramètre du modele {clf7_entropy.get_params()}")
print(f"entropy_errors {entropy_errors}")
```

```python
#gini
idmax =20
max_depths = range(1,idmax)

pgrid = {"max_depth": max_depths,"min_samples_split": [2, 3, 5, 10, 15, 20]}
gini_grid_search = GridSearchCV(clf7_gini, param_grid=pgrid, cv=5)
def f(a):
    return gini_grid_search.predict(a.reshape(-1,1))
plt.figure(1, figsize=(15, 12))
gini_grid_search.fit(X_train,y_train)
plt.subplot(2,1,1)
plt.title('gini')
frontiere_new(f,X_test,y_test)
print(f"gini_best_estimator {gini_grid_search.best_estimator_.score(X_test,y_test)}")
gini_grid_search.best_params_
# -->{'max_depth': 2}


# entropy
entropy_grid_search = GridSearchCV(clf7_entropy, param_grid=pgrid, cv=5)
def f(a):
    return entropy_grid_search.predict(a.reshape(1,-1))
plt.figure(1, figsize=(15, 12))
entropy_grid_search.fit(X_train,y_train)
plt.subplot(2,1,2)
plt.title('entropy')
frontiere_new(f,X_test,y_test)
print(f"entropy_best_estimator {entropy_grid_search.best_estimator_.score(X_test,y_test)}")
entropy_grid_search.best_params_
```

```python
data_test = rand_checkers(40, 40, 40, 40, sigma=0.1)
X_test = data_test[:, :2]
Y_test = data_test[:, 2].astype(int)
```

```python
dmax = 12
scores_entropy = np.zeros(dmax)
scores_gini = np.zeros(dmax)
# Plot separation rules using entropy
print("Separation rules using entropy as impurity criterion:")
plt.close(2)
plt.figure(2, figsize=(15, 10))
for i in range(dmax):
    dt_entropy = tree.DecisionTreeClassifier(criterion='entropy', max_depth=i + 1)
    dt_entropy.fit(X5, y5)
    scores_entropy[i] = dt_entropy.score(X_test, Y_test)
    plt.subplot(3, 4, i + 1)
    frontiere_new(lambda x: dt_entropy.predict(x.reshape((1, -1))), X_test, Y_test, step=50, samples=False)
plt.show()
plt.close(3)
print(" ")
# Plot separation rules using Gini index
print("Separation rules using Gini index as impurity criterion:")
plt.figure(3, figsize=(15, 10))
for i in range(dmax):
    dt_gini = tree.DecisionTreeClassifier(criterion='gini', max_depth=i + 1)
    dt_gini.fit(X5, y5)
    scores_gini[i] = dt_gini.score(X_test, Y_test)
    plt.subplot(3, 4, i + 1)
    # Set 'samples=True' to plot the data along with the frontier
    frontiere_new(lambda x: dt_gini.predict(x.reshape((1, -1))), X_test, Y_test, step=50, samples=False)
plt.show()
plt.close(4)
# Summarize the results:
#  - in print
print('Entropy produced scores: {}'.format(scores_entropy))
print('Gini-index produced scores: {}'.format(scores_gini))
#  - on graphics
plt.figure(4)
plt.plot(np.arange(1, dmax + 1), scores_entropy, label='Entropy')
plt.plot(np.arange(1, dmax + 1), scores_gini, label='Gini index')
plt.title('Accuracy on test data')
plt.xlabel('Max depth')
plt.ylabel('Accuracy Score')
plt.legend()
```
